### Series - Basic indexing

```python
import pandas as pd
import numpy as np

x = pd.Series(np.random.normal(size = 10**7)) # 10 million
print(x)
print(x.index)

# 10 million de nuevo.
# El index es de tipo Int64, consume más memoria pq almacena todos los valores del indice, mientras que el anterior solo almacena 3 valores, start, stop, step.
# El primero es más rápido en encontrar un valor dado su indice ex 56123 ya que es un RangeIndex y ese valor estará en la posición esperable
# En el segundo no hay garantia de que el valor 56123 esté en el lugar esperado dado el valor del indice ya que el indice puede estar ordenado de cualquier forma.
# Tampoco hay garantía de que el segundo no tenga duplicados.
y = pd.Series(np.random.normal(size = 10**7), index = np.arange(10**7))  
print(y)
print(y.index)

import sys
print(sys.getsizeof(x))
print(sys.getsizeof(y))


# Ejemplo con repetidos
alpha = pd.Series([2,3,5, 7, 11], index = [0,0,1,1,2])
print(alpha)
print(alpha[0])
print(alpha.loc[0])
print(alpha.iloc[0]) # Este sabe que queremos solo el indice

# Multi write
z = pd.Series([10, 20, 30, 40, 50], index = ['a','b','c','d','e'])
print(z)
print(z.iloc[1])
z.iloc[1] = 200
print(z.iloc[1])
z.iloc[[0,1,2]] = 99
print(z)
z.iloc[:3] = 999 # all but pos 3
print(z)

# Overwrite with new values, share some common index values
z.iloc[:] = np.array([5, 10, 20, 30, 40])

# 2 pandas, same indexes

za = pd.Series([10, 20, 30, 40])
zb = pd.Series([11, 22, 33, 44], index = [7, 3, 2, 0])
print(za)
print(zb)

# Pandas start searching zA for values with index 0, 1. 
# Looks for matching values in zB for overriding those, if not found, NaN. 
# Also cast everything to float to use nan.
za.loc[ [0,1] ] = zb
print(za)
print(zb)

za.loc[:2] = zb
print(za)

```

### Basic Operations

```python
# Panda uses broadcasting
import pandas as pd
import numpy as np

x = pd.Series([1,2,3,4])
print(x)

x2 = x + 1 # Pandas uses broadcasting, it adds 1 to every element of the series
print(x2)

y = pd.Series(1)
xy = x + y # If you combine two series, pandas only combines elements with the same index. Index alignment.
print(xy)

A = pd.Series([10,20,30,40,50])
B = pd.Series([1,2,3,4,5])

print(A+B) # Adition happens element wise by index, not by position, if we reverse an index we can confirm it

AA = pd.Series([10,20,30,40,50], index = [4,3,2,1,0])
BB = pd.Series([1,2,3,4,5])
print(AA+BB)

# The same for every other basic math operation. For more complex operations we use numpy.

AABB = AA.to_numpy() + BB.to_numpy()
print(AABB)
print(pd.Series(AABB))

# We can mix numpy and series

print(A + B.to_numpy()) # A is keeping it indexes while B no

# We want to add Y to A on matching indexes without Nan on not founds
Y = pd.Series([10, 20], index = [2,1])

# Match only y indexes -> x.loc[y.index]
# Then add y, it will ignore unmatched indexes += Y
x.loc[Y.index] += Y 
print (x)

```

# Boolean indexing

```python
import pandas as pd
import numpy as np

x = pd.Series([20,50,11,45,17,31])
print(x)

mask = x < 20
print(mask)

print(x.loc[mask]) # subset of x where mask is true

x.index = [0, 1, 2, 3, 5, 4]
print(x.loc[mask]) # It is affected by index changes
print(x.loc[mask.to_numpy()]) # It is NOT affected by index changes

# You can combine boolean filter with and &, or |
mask2 = (x > 5) & (x < 30)
print(mask2)
print(x.loc[mask2])

# We can negate masks
print(x.loc[~mask2])
```

# Missing values

```python
import pandas as pd
import numpy as np

roux = pd.Series([1, 2, 3]) # it is int
print(roux)

roux.iloc[1] = np.nan # Now it is float
print(roux)

# Podamos usar un nullable de pandas
roux = pd.Series([1, 2, 3], dtype = 'Int64') # it is Int64
roux.iloc[1] = np.nan # Keeps it, it is not casted
print(roux)

# We can also do
roux = pd.Series([1, 2, 3], dtype = 'Int64')
roux.iloc[1] = pd.NA 
print(roux)

# With strings
gumbo = pd.Series(['a', 'b', 'c'])
gumbo.iloc[1] = np.nan
gumbo.iloc[2] = None
print(gumbo) # It is a python list, a list of pointer to objects. Also distinct objects: (

gumbo = pd.Series(['a', 'b', 'c'], dtype='string') # Fixed with dtype string
gumbo.iloc[1] = np.nan
gumbo.iloc[2] = None
print(gumbo)

# Check for NA
maskNa = pd.isna(roux)
maskNotNa = pd.notna(roux)
print(maskNa)
print(maskNotNa)

y = roux.fillna(-2) # returns a modified copy, unless inplace = True
print(y)
# Or

roux.loc[maskNa] = -1
print(roux)
```

# Use apply

```python
foo = pd.Series([3, 9, 2, 2, 8, 7])

def my_fn(x):
    return x - 1 if x % 2 == 0 else x + 3

print(foo.apply(my_fn))

def my_fn2(x, a, b):
    return x - 1 if x % 2 == 0 else x + 3

print(foo.apply(my_fn2, a = 3, b = -4))

# This is slow :(, we can do this on numpy, faster
a = foo.to_numpy()
y = pd.Series(np.where(a %2 == 0, a-1, a+3)) # esto es un if inline funciona parece
print(y)
```

# View vs Copy

```python
import pandas as pd
import numpy as np

x = pd.Series([2, 3, 5, 7, 11], index = [2, 11, 12, 30, 30])
y = x
y.iloc[0] = 9 # Reference 
print(x)
print(y)

y = x.copy()
y.iloc[0] = 999
print(x)
print(y)

foo = pd.Series(['a', 'b', 'c', 'd'], dtype = 'string')
print(foo)

bar = foo.loc[foo <= 'b']  # does not affect foo
print(bar)
bar.iloc[0] = 'x'

print(foo)
print(bar)

bar = foo.iloc[:2] # does affect foo
print(bar)
bar.iloc[0] = 'x' 
print(foo)
print(bar)

# Using a = b.loc -> returns a copy. Otherwise it returns a view. This is undocumented :(. Use .copy to be safe

# Any pandas function that modifies a series, ex replace
xoo = pd.Series(['tiger', 'lion', 'zebra', 'lion'])
print(xoo)
xoo2  = xoo.replace({'lion': 'cat', 'tiger': 'fast cat'}) # returns a changed copy unless you do xoo = xoo.replace ... or inplace = True
print(xoo)
print(xoo2)
```

# Apply


```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [5.2, 1.7, 9.4],
    'B': [3.9, 4.0, 7.8]
})

print(df)
# Args -> func (fn)
# Axis -> Axis 0 row, or axis 1 column.

print(df.apply(func = np.sum, axis = 1))
print(df.apply(func = np.sum, axis = 0))

kids = pd.DataFrame({
    'name': pd.Series(['alice', 'mary', 'jimmy', 'johny', 'susan'], dtype='string'),
    'age': [9, 13, 11, 15, 8],
    'with_adult': [True, False, False, True, True]
})
print(kids)

# At least 12 or have aduct suppervision
def is_allowed(kids_series):
    return kids_series.loc['age'] >= 12 or kids_series.loc['with_adult']

print(kids.apply(is_allowed, axis = 1))

def is_allowed_clean(age, with_adult):
    return age >= 12 or with_adult

print(kids.apply(lambda row: is_allowed_clean(row.loc['age'], row.loc['with_adult']), axis = 1))
```

```python
import pandas as pd
import numpy as np

# https://www.practiceprobs.com/problemsets/python-pandas/dataframe/#merging-dataframes
pets = pd.DataFrame(
    data={
        'name': ['Mr. Snuggles', 'Honey Chew Chew', 'Professor', 'Chairman Meow', 'Neighbelline'],
        'type': ['cat', 'dog', 'dog', 'cat', 'horse']
    },
    index=[71, 42, 11, 98, 42]
)

visits = pd.DataFrame(
    data={
        'pet_id': [42, 31, 71, 42, 98, 42],
        'date': ['2019-03-15', '2019-03-15', '2019-04-05', '2019-04-06', '2019-04-12', '2019-04-12']
    }
)

print(pets)
#                name   type
# 71     Mr. Snuggles    cat
# 42  Honey Chew Chew    dog
# 11        Professor    dog
# 98    Chairman Meow    cat
# 42     Neighbelline  horse

print(visits)
#    pet_id        date
# 0      42  2019-03-15
# 1      31  2019-03-15
# 2      71  2019-04-05
# 3      42  2019-04-06
# 4      98  2019-04-12
# 5      42  2019-04-12

# inner join between pets and visits based on pet_id
pd.merge(
    left=pets, 
    right=visits, 
    how='inner', 
    left_index=True, 
    right_on='pet_id'
)
#               name   type  pet_id        date
# 2     Mr. Snuggles    cat      71  2019-04-05
# 0  Honey Chew Chew    dog      42  2019-03-15
# 3  Honey Chew Chew    dog      42  2019-04-06
# 5  Honey Chew Chew    dog      42  2019-04-12
# 0     Neighbelline  horse      42  2019-03-15
# 3     Neighbelline  horse      42  2019-04-06
# 5     Neighbelline  horse      42  2019-04-12
# 4    Chairman Meow    cat      98  2019-04-12
# 

pets.index.rename('pet_id', inplace=True)

print(pets)
#                    name   type
# pet_id                        
# 71         Mr. Snuggles    cat
# 42      Honey Chew Chew    dog
# 11            Professor    dog
# 98        Chairman Meow    cat
# 42         Neighbelline  horse

visits.index.rename('visit_id', inplace=True)

print(visits)
#           pet_id        date
# visit_id                    
# 0             42  2019-03-15
# 1             31  2019-03-15
# 2             71  2019-04-05
# 3             42  2019-04-06
# 4             98  2019-04-12
# 5             42  2019-04-12

pd.merge(left=pets, right=visits, how='inner', on='pet_id')
#    pet_id             name   type        date
# 0      71     Mr. Snuggles    cat  2019-04-05
# 1      42  Honey Chew Chew    dog  2019-03-15
# 2      42  Honey Chew Chew    dog  2019-04-06
# 3      42  Honey Chew Chew    dog  2019-04-12
# 4      42     Neighbelline  horse  2019-03-15
# 5      42     Neighbelline  horse  2019-04-06
# 6      42     Neighbelline  horse  2019-04-12
# 7      98    Chairman Meow    cat  2019-04-12

pd.merge(left=pets, right=visits, how='left', on='pet_id')
#    pet_id             name   type        date
# 0      71     Mr. Snuggles    cat  2019-04-05
# 1      42  Honey Chew Chew    dog  2019-03-15
# 2      42  Honey Chew Chew    dog  2019-04-06
# 3      42  Honey Chew Chew    dog  2019-04-12
# 4      11        Professor    dog         NaN
# 5      98    Chairman Meow    cat  2019-04-12
# 6      42     Neighbelline  horse  2019-03-15
# 7      42     Neighbelline  horse  2019-04-06
# 8      42     Neighbelline  horse  2019-04-12

pd.merge(left=pets, right=visits, how='right', on='pet_id')
#    pet_id             name   type        date
# 0      42  Honey Chew Chew    dog  2019-03-15
# 1      42     Neighbelline  horse  2019-03-15
# 2      31              NaN    NaN  2019-03-15
# 3      71     Mr. Snuggles    cat  2019-04-05
# 4      42  Honey Chew Chew    dog  2019-04-06
# 5      42     Neighbelline  horse  2019-04-06
# 6      98    Chairman Meow    cat  2019-04-12
# 7      42  Honey Chew Chew    dog  2019-04-12
# 8      42     Neighbelline  horse  2019-04-12

```

# Groupby

```python
df = pd.DataFrame({
    'A': ['foo', 'bar', 'foo', 'bar', 'bar', 'foo', 'foo'],
    'B': [False, True, False, True, True, True, True],
    'C': [2.1, 1.9, 3.6, 4.0, 1.9, 7.8, 2.8],
    'D': [50, np.nan, 30, 90, 10, np.nan, 10]
}).convert_dtypes()  

print(df)

x = df.groupby( by = 'A' ) # Partition by diff values on A
print(x)


x = df.groupby( by = ['A', 'B'] ) # Partition by diff values on A,B
print(x.groups)
print(x.ngroup())

# Creamos un valor calculado, mantiene el indicie y el número de valores
wholes = df.C.apply(np.floor)
print(wholes)
# Agrupamos por el valor calculado, lo hace gracias al indice
by_wholes = df.groupby(wholes)
print(by_wholes.groups)

y = df.groupby('A')['C'].sum()
# y = df.groupby('A') <- Dataframe groupby object 
# y = df.groupby('A')['C'] <- Isolate column C on the grouping
# y = df.groupby('A')['C'].sum() -> Sum of C by groups (series)
print(y)

# Pandas a series of series, allow to grab value by col key
y = df.groupby('A')[['C']].sum()
print(y)

y = df.groupby('A')[['C', 'D']].sum()
print(y)

y = df.groupby('A')[['C', 'D']].agg('sum')
print(y)

y = df.groupby('A')[['C', 'D']].agg({'C': np.sum, 'D': ['mean', 'var']})
print(y)

y = df.groupby(['A', 'B'])[['D']].mean()
print(y)

y = df.groupby(by=['B', 'D'], dropna=False).agg({'C':['min', 'max']})
print(y)

y = df.groupby(by=['B', 'D'], dropna=False).agg(
    C_min=('C', 'min'),
    C_max=('C', np.max)
)
print(y)

y = df.groupby('A')['C'].transform(lambda x: x.sort_values())
y = df.groupby('A')['C'].transform(lambda x: pd.Series.sort_values)
print(y)

df['C_sorted_within_A'] = df.groupby('A')['C'].transform(pd.Series.sort_values)
print(df)

df.groupby('A')[['C', 'D']].transform('mean')
```
